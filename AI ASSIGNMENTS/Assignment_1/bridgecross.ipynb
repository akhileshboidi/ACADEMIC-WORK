```python
class State:
    def __init__(self,left,right,umbrella,time):
        self.left=left
        self.right=right
        self.umbrella=umbrella
        self.time=time

    def goalTest(self):
        if(len(self.left)==0 and self.umbrella=="R" and self.time<=60):
            return True
        else:
            return False

    def moveGen(self):
        children=[]
        people_times={"A":5,"B":10,"C":20,"D":25}

        if self.umbrella=='L':
            #move 2 people or 1 person from left to right
            for i in range(len(self.left)):
                for j in range(i,len(self.left)):
                    p1=self.left[i]
                    p2=self.left[j]
                    new_left=self.left.copy()
                    new_left.remove(p1)
                    if(p1!=p2):
                        new_left.remove(p2)
                        time_taken=max(people_times[p1],people_times[p2])
                        new_right=self.right+[p1,p2]
                    else:
                        time_taken=people_times[p1]
                        new_right=self.right+[p1]
                    child_obj=State(new_left,new_right,"R",self.time+time_taken)
                    children.append(child_obj)

        else:
                #umbrella is in right so move 1 person and umbrella from right to left
            for p in self.right:
                new_right=self.right.copy()
                new_right.remove(p)
                new_left=self.left+[p]
                time_taken=people_times[p]
                child_obj=State(new_left,new_right,"L",self.time+time_taken)
                children.append(child_obj)

        return children
    

    def __eq__(self, other):
        return (
            sorted(self.left) == sorted(other.left) and
            sorted(self.right) == sorted(other.right) and
            self.umbrella == other.umbrella 
        )

    def __hash__(self):
        return hash((tuple(sorted(self.left)), tuple(sorted(self.right)), self.umbrella, self.time))

    
    
    def __str__(self):
        return f"L:{sorted(self.left)} ,R:{sorted(self.right)} ,Umbrella:{self.umbrella} ,Time:{self.time}"

        

def reconstructpath(node_pair,closed):
    path=[]
    parent_map={}
    for node,parent in closed:
        parent_map[node]=parent

    node,parent=node_pair
    path.append(node)
    while parent is not None:
        path.append(parent)
        parent=parent_map[parent]

    return path


def removeseen(children,open,closed):
    open_nodes=[node for node,parent in open]
    closed_nodes=[node for node,parent in closed]
    new_nodes=[c for c in children if c not in open_nodes and c not in closed_nodes]
    return new_nodes


def bfs(start):
    open=[(start,None)]
    closed=[]
    while open:
        node_pair=open.pop(0)
        node,parent=node_pair
        if(node.goalTest()):
            print("Goal Found")
            path=reconstructpath(node_pair,closed)
            path.reverse()
            for p in path:
                print("->" ,p)
            print("\nMinimum Time taken to cross the Bridge:", node.time, "minutes")
            return
                
        else:
            closed.append(node_pair)
            children=node.moveGen()
            new_nodes=removeseen(children,open,closed)
            new_pairs=[(c,node) for c in new_nodes]
            open=open+new_pairs
    print("No Solution found within 60 minutes")
    return []

def dfs(start):
    open=[(start,None)]
    closed=[]
    while open:
        node_pair=open.pop(0)
        node,parent=node_pair
        if(node.goalTest()):
            print("Goal Found")
            path=reconstructpath(node_pair,closed)
            path.reverse()
            for p in path:
                print("->" ,p)
            print("\nMinimum Time taken to cross the Bridge:", node.time, "minutes")
            return
                
        else:
            closed.append(node_pair)
            children=node.moveGen()
            new_nodes=removeseen(children,open,closed)
            new_pairs=[(c,node) for c in new_nodes]
            open=new_pairs+open
    print("No Solution found within 60 minutes")
    return []



start_state=State(left=['A' ,'B','C','D'],right=[],umbrella='L',time=0)
print("Bridge Crossing BFS:")
bfs(start_state)
print("Bridge Crossing DFS:")
dfs(start_state)
```
